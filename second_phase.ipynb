```python
from src import *
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## Modelo 5

```python
def model5():
    model=Sequential()
    model.add(Dense(128, input_dim=12, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(16, activation='linear'))
    return model
```

```python
agent = Agent(learning_rate=1e-4, map_file="0o100t100s1p100", num_clusters=4, pre_train=True,
              name=f"models/m5_n4_p2_min", model=model5)
```

```python
episodios=50
episodios_terminados=0
limit=150
```

```python
while episodios_terminados < 1000:
    agent.train(episodes=episodios, finished_episodes=episodios_terminados, batch_size=32, epsilon=1,
                epsilon_decay=0.997, score_limit=limit, refresh_rate=2, verbose=False, 
                block_impossible_actions=True, log=True, run_test=False)
    agent.save(f"models/m5_n4_p2_{episodios+episodios_terminados}")
    episodios_terminados += episodios
agent.rewards_graphic(n=10, ylim=limit+1)
```
