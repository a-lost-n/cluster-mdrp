```python
from src import *
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
episodios=1000
episodios_terminados=0
decay=0.997
limit=150
```

```python
agent = Agent(learning_rate=1e-4, map_file="0o100t100s1p100", num_clusters=4, name="models/m2_n4")
agent.train(episodes=episodios, finished_episodes=episodios_terminados, batch_size=32, epsilon=1,
             epsilon_decay=decay, score_limit=limit, refresh_rate=2, verbose=False, 
             display_rewards=False, block_impossible_actions=True)
```

```python
agent.map.display_map_clusters()
```

## Modelo 1
Input(12)->ReLU(576)->ReLU(192)->ReLU(64)->Linear(16)

```python
def model1():
    model=Sequential()
    model.add(Dense(576,input_dim=12, activation='relu'))
    model.add(Dense(192, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(16, activation='linear'))
    return model
```

```python
agent = Agent(map_file="0o100t100s1p100", filename="models/m1_n4", num_clusters=4, pre_train=False, model=model1)
agent.test(episodes=200, score_limit=250)
```
